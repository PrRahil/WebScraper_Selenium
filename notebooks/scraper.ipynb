```python
import sys
import os

# Add src folder to system path so we can import it
sys.path.append(os.path.join(os.getcwd(), '../src'))

from scraper import WebScraper

# Create an instance of the scraper
scraper = WebScraper(driver_path='../chromedriver.exe')

# Step 1: Open a website
url = 'https://abdm.gov.in/dhis'
scraper.open_website(url)

# Step 2: Extract all text content
all_text = scraper.extract_all_text()
print("All Text Content:", all_text)

# Step 3: Find and download files
downloaded_files = scraper.find_and_download_files()
print("Downloaded Files:", downloaded_files)

# Step 4: Collect the data in a structured way
scraped_data = {
    "url": url,
    "text_content": all_text,
    "downloaded_files": downloaded_files
}

# Step 5: Save the data to a JSON file in the data folder
scraper.save_data_to_json(scraped_data)

# Step 6: Close the browser
scraper.close_browser()
```
